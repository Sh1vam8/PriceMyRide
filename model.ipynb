```python
X = df[['insurance_validity','fuel_type','kms_driven','ownsership','transmission']] 
y = df['price(in lakhs)']
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
models = {'LinReg':LinearRegression(),
         'KNN-3':KNeighborsRegressor(n_neighbors = 3),
         'SVN-Poly':SVR(kernel='poly')}
for name,model in models.items():
    model.fit(X,y)
    sc= model.score(X,y)
    print(name,sc)
```

```python
final_model = models['KNN-3']
```

```python
k1 = {'Comprehensive':0,'Third Party insurance':1,'Zero Dep':2,'Not Available':3,'Third Party':1}
k2 = {'Petrol':1,'Diesel':2,'CNG':3}
k3 = {'First Owner':1,'Second Owner':2,'Third Owner':3,'Fifth Owner':5}
k4 = {'Manual':7,'Automatic':8}
```

```python
insurance_validity = input('Insurance_validity:')
fuel_type = input('Fuel_type:')
kms_driven = input('Kms Driven:')
ownsership = input('Ownership:')
transmission = input('Transmission Type:')
insurance_validity = k1[insurance_validity]
fuel_type = k2[fuel_type]
ownsership = k3[ownsership]
transmission = k4[transmission]
kms_driven = int(kms_driven)
test = [[insurance_validity,fuel_type,kms_driven,ownsership,transmission]]
yp = int(final_model.predict(test)[0])
print('Predicted Price of the Car is {} $$.'.format(yp))
```

```python
import pickle
pickle.dump(final_model,open('final_model.pkl','wb'))
```
